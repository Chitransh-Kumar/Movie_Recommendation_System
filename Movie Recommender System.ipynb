```python
# Merging both the datasets together and storing it in movies dataset

movies=movies.merge(credits,on='title')
```

In my machine learning model, I have considered the following features of the dataset to be useful to recommend movies.
The features are:
1. genres
2. id
3. keywords
4. title
5. overview
6. cast
7. crew

```python
# Only the above-mentioned columns are considered for future model building

movies=movies[['genres','id','keywords','title','overview','cast','crew']]
```

```python
# Overview of the dataset

movies.head()
```

```python
# Checking for null values

movies.isnull().sum()
```

```python
movies.shape
```

```python
# Since the null values are very less compared to the total size, these values are dropped from the dataset

movies.dropna(inplace=True)
```

```python
movies.shape
```

```python
# Checking for duplicate values in the dataset

movies.duplicated().sum()
```

```python
# Importing ast library to parse string to dictionary

import ast
```

```python
# Defining a function to get only genre values from 'genres' feature in the dataset

def get_genre(obj):
    genre=[]
    for i in ast.literal_eval(obj):
        genre.append(i['name'])
    return genre
```

```python
# Applying the above defined function to 'genres' feature in the dataset

movies['genres']=movies['genres'].apply(get_genre)
```

```python
# Overview of the dataset

movies.head()
```

```python
# Defining a function to get keyword values from 'keywords' feature in the dataset

def get_keywords(obj):
    keywords=[]
    for i in ast.literal_eval(obj):
        keywords.append(i['name'])
    return keywords
```

```python
# Applying the above function to 'keywords' feature in the dataset

movies['keywords']=movies['keywords'].apply(get_keywords)
```

```python
# Overview of the dataset

movies.head()
```

```python
# Defining a function to get top 3 cast (actors/actresses) from 'cast' feature in the dataset

def get_cast(obj):
    cast=[]
    cnt=0
    for i in ast.literal_eval(obj):
        if cnt!=3:
            cast.append(i['name'])
            cnt+=1
        else:
            break
    return cast
```

```python
# Applying the above function to 'cast' feature in the dataset

movies['cast']=movies['cast'].apply(get_cast)
```

```python
# Overview of the dataset

movies.head()
```

```python
# Defining a function to get only director's name from 'crew' feature of the dataset

def get_director(obj):
    director=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            director.append(i['name'])
            break
    return director
```

```python
# Applying the above function to 'crew' feature of the dataset

movies['crew']=movies['crew'].apply(get_director)
```

```python
# Overview of the dataset

movies.head()
```

```python
# Applying lambda function in 'overview' feature of the dataset to split each word of the paragraph

movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
# Overview of the dataset

movies.head()
```

```python
# Now the idea is if there are two separate words representing the same value, it would be bad for ML model.
# Eg: drug abuse: It can be considered as two separate words 'drug' and 'abuse'
# Therefore, we need to merge both the words into single word.
# Eg: drug abuse changes to drugabuse.
```

```python
# Therefore, we apply this idea to all the required features (genres, keywords, cast, crew) of the dataset.

features=['genres', 'keywords', 'cast', 'crew']

for feature in features:
    movies[feature]=movies[feature].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# Overview of the dataset

movies.head()
```

```python
# Merging 'overview', 'genres', 'keywords', 'cast' and 'crew' into single feature 'tag'.

movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
# Overview of the dataset

movies.head()
```

```python
# Dropping the unnecessary features from the dataset

movies.drop(columns=['genres','keywords','overview','cast','crew'],inplace=True)
```

```python
# Overview of the dataset

movies.head()
```

```python
# Converting content in 'tag' feature from list to string

movies['tag']= movies['tag'].apply(lambda x:" ".join(x))
```

```python
# Overview of the dataset

movies.head()
```

```python
# Converting content in 'tag' feature to lowercase characters

movies['tag']= movies['tag'].apply(lambda x:x.lower())
```

```python
# Overview of the dataset

movies.head()
```

```python
# Importing PorterStemmer NLP Library to consider all the similar words as same

from nltk.stem.porter import PorterStemmer
```

```python
# Creating class of the object

ps= PorterStemmer()
```

```python
# Defining a function to consider all the similar words as same in 'tag' feature

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
# Applying the function to 'tag' feature

movies['tag']= movies['tag'].apply(stem)
```

```python
# Importing CountVectorizer Library to convert values in 'tag' into vectors

from sklearn.feature_extraction.text import CountVectorizer 
```

```python
# Creating class of the object with max_features as 5000 and language to be considered as English

cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
# Creating vectors

vectors=cv.fit_transform(movies['tag']).toarray()
```

```python
# Importing consine_similarity library to calculate cosine distance

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Creating similarity variable to calculate cosine distance

similarity=cosine_similarity(vectors)
```

```python
# Main 'Recommend' function for movie recommendation

def recommend(movie):
    movie_index= movies[movies['title']== movie].index[0]
    distance= similarity[movie_index]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
# Testing the function for 'Avatar' movie

recommend('Avatar')
```

```python
# Testing the function for 'Pirates of the Caribbean: At World's End' movie

recommend("Pirates of the Caribbean: At World's End")
```

```python
# Testing the function for 'Spectre' movie

recommend('Spectre')
```
